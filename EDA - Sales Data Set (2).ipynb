# DATA VISUALIZATION - BARPLOTTING

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("whitegrid")
Snapshot_of_Product_Cost = sns.barplot(x="Year", y="Product cost", data=df)
plt.title("Product Cost")
```

```python
sns.set_style("whitegrid")
Snapshot_of_Gross_Profits = sns.barplot(x="Year", y="Gross profit", data=df)
plt.title("Gross Profits")
```

```python
sns.set_style("whitegrid")
Snapshot_of_Revenue_Earnings = sns.barplot(x="Year", y="Revenue", data=df)
plt.title("Revenue Earnings")
```

```python
#bar plot revenue comparison for phone orders for camp gear in the US from 2004-2007
df = pd.DataFrame({
    "Year": ["2004", "2005", "2006", "2007"],
    "Revenue for camp gear": [315044.33, 253305.72, 19756.56, 0],
    "Predicted revenue for camp gear": [437477.15, 289327.36, 20798.04, 0]})
df.plot.bar() 
plt.bar(df['Predicted revenue for camp gear'], df['Revenue for camp gear']) 
plt.xlabel("Year") 
plt.ylabel("Revenue in USD") 
plt.show() 
```
